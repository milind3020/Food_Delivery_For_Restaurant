```python
import json
import sys
import datetime
import os



class Admin:
    
    def __init__(self):
        
        self.food_item={}
    
        
    def generate_id(self,string):
        id = 0
        for char in string:
            id += ord(char)
        return int(id)   
        
    
    def admin_login(self):
        self.admin_username=input("Enter Admin Username: ")
        self.admin_password=input("Enter Admin Password: ") 
        if self.admin_username=='admin' and self.admin_password=="admin":
            self.admin_menu()
        else:
            print("\n" * 10 + "ERROR: Invalid Input  Try again!")
            self.admin_login()
            
    def admin_menu(self):
        while True:
            
                print("*" * 31 + "Choose One Option" + "*" * 32 + "\n"     
              "\t(A) ADD ITEMS\n"                              
              "\t(U) UPDATE ITEMS\n"
              "\t(V) VIEW ITEMS\n"
              "\t(R) REMOVE ITEMS\n"
              "\t(E) EXIT\n" +
              "_" * 72)
                input_1 = str(input("Please Select Your Operation: ")).upper()
                if (len(input_1) == 1):                                           
                    if (input_1 == 'A'):                                          
                        print("\n" * 2)                                        
                        self.get_food_item()                          
                        break                                                     
                    elif (input_1 == 'U'):                                        
                        print("\n" * 2) 
                        self.edit_item()
                        break                                                     
                    elif (input_1 == 'V'):                                        
                        print("\n" * 2)
                        self.view_item()
                        break 
                    elif (input_1 == 'R'):                                        
                        print("\n" * 2)
                        self.del_item()
                        break                                                    
                    elif (input_1 == 'E'):                                        
                        print("*" * 32 + "THANK YOU" + "*" * 31 + "\n")           
                        break                                                     
                    else:                                                                                 
                        print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!")     
                else:                                                                                     
                    print("\n" * 10 + "ERROR: Invalid Input (" + str(input_1) + "). Try again!")
        
        
    def get_food_item(self):
        try:
            with open("Food_item.json","r") as f:
                self.food_item=json.load(f)
        except FileNotFoundError:
            self.food_item={}
        self.name=input("Enter Food Name: ")
        self.quantity=(input("Enter quantity"))
        self.price=(input("Enter Price: "))
        self.discount=(input("Enter Discount: "))
        self.stock=(input("Enter Stock: ")) 
        self.item={"Foodname": self.name,"Quantity":self.quantity,"Price":self.price,"Discount":self.discount,"Stock":self.stock}
        #self.id=self.generate_food_id()
        self.food_item[self.generate_id(self.name)]=self.item
        with open("Food_item.json","w") as f:
            json.dump(self.food_item,f)
            print("Added Successfully!!!")
        self.admin_menu()
            
        return self.food_item
    def del_item(self):
        if os.path.getsize("Food_item.json") > 0:
            with open("Food_item.json","r") as f:
                out_data=json.load(f)
            if out_data:
                for k,v in out_data.items():
                    print("Id:",k)
                    for a in v:
                        print(a,":",v[a])
            else:
                print("There are No items please add items first I am sending you to Main Menu")
                self.admin_menu()
        id1=(input("Enter food id"))
        del out_data[id1]
        with open("Food_item.json","w") as f:
            json.dump(out_data,f)
        print("Delted Successfully")
        x=input("********Press Enter To Go Back To Main Menu********")
        if x=="":
            self.admin_menu()
        
    def view_item(self):
        with open("Food_item.json","r") as f:
            out_data=json.load(f)
        if out_data:
            for k,v in out_data.items():
                print("Id:",k)
                for a in v:
                    print(a,":",v[a])
        else:
            
            print("***********There are no food items added please add food items first****************")
        x=input("********Press Enter To Go Back To Main Menu********")
        if x=="":
            self.admin_menu()
    def edit_item(self):
        
        with open("Food_item.json",'r') as f:
            o_d=json.load(f)
            
            
        for k,v in o_d.items():
            print("Id:",k)
            for a in v:
                print(a,":",v[a])
        food_item=(input("Enter id of food item to edit:"))
        x=input("Enter the Field to update").capitalize()
        o_d[food_item][x]=input("Enter the new value ")
        with open("Food_item.json","w") as f:
            json.dump(o_d,f)
        print("Field updated Successfully")
        
        x=input("********Press Enter To Go Back To Main Menu********")
        if x=="":
            self.admin_menu()
            
###################################################################################################
class User:
    def __init__(self):
        self.user_details={}
        
    
        
    def reg_form(self):
        try:
            with open("reg_users.json","r") as f:
                self.user_details=json.load(f)
        except FileNotFoundError:
            self.user_details={}
        self.name=input("Enter Your Full Name:")
        self.mo_no=(input("Enter Your Phone No : "))
        self.email=input("Enter Email Id: ")
        self.addr=input("Enter Your Address: ")
        self.passw=input("Enter your password: ")
        self.item={"Name":self.name,"Mobile":self.mo_no,"Email":self.email,"Address":self.addr,"Password":self.passw}
        self.user_details[self.email]=self.item
        with open("reg_users.json","w") as f:
            json.dump(self.user_details,f)
        print("User Registered Successfully!!!!!!!!!!!")
        main_screen()
    def user_login(self):
        email=input("Enter Registered Email ID: ")
        passw=input("Enter Your password")
        try:
            with open("reg_users.json","r") as f:
                data=json.load(f)
        except:
                print("You ARe not registerd pleasde register yourself first!!!!")
                sys.exit()
       
        for k,v in data.items():
            if k==email and v["Password"]==passw:
                self.email_f=email
                self.user_menu()
                
            else:
                print("Wrong Credentials Entered")
       
    
    def user_menu(self):
        print('---------------------------')
        print('|     (P)Place New Order   |')
        print('---------------------------')
        print('---------------------------')
        print('|     (O)Order History     |')
        print('---------------------------')
        print('---------------------------')
        print('|     (U) Update Profile  |')
        print('---------------------------')
        while True:
            print("NOTE: To Exit Just Enter E")
            ch=input("Please Input Your Choice").upper()
            if (len(ch)==1):
                if ch=='P':
                    self.place_order()
                    break
                elif ch=="O":
                    self.order_history()
                    break
                elif ch=="U":
                    self.update_regis() 
                    break
                elif ch=="E":
                    print("*" * 32 + "THANK YOU" + "*" * 31 + "\n") 
                    sys.exit()
                else:                                                                                 
                    print("\n" * 10 + "ERROR: Invalid Input . Try again!")   
            else:                                                                                     
                print("\n" * 10 + "ERROR: Invalid Input . Try again!")

        
    
    

    def place_order(self):
    
        with open("Food_item.json","r") as f:
            content=f.read()
            data=json.loads(content)
        
        li=[]
        for k,v in data.items():
            x=(list(v.values()))
            
            li.append(x)
            
        for i,z in enumerate(li,1):
            print(f"[{i}] {z[0].capitalize()} ({z[1]}) [INR{z[2]}]")
        x=set(map(int,input("Enter Items to order separated by commas ").split(",")))
        sorted(x)
        f_l=[]
        for i in x:
            f_l.append(li[i-1][:3])
        print("This is Your Final Order List")
        for item in f_l:
            print(f"{item[0].capitalize()} ({item[1]}) [INR{item[2]}]", end=' ')
            print()
        l_in=input("Enter CONFIRM to place your order").upper()
        if l_in=="CONFIRM":
            file_name = self.email_f + ".json"
            try:
                with open(file_name,"r") as f:
                    order_dict=json.load(f)
            except FileNotFoundError:
                    order_dict={}
        
        curr=datetime.datetime.now()
        time_key=curr.strftime("%Y-%m-%d %H:%M")
        order_dict[time_key]=f_l
            
        with open(file_name,"w") as f:
            json.dump(order_dict,f)
        print("Order Successfull!!!!!!!")
        ch=input("****Press Enter To go back to main menu****")
        if ch=="":
            self.user_menu()
            
        
    def update_regis(self):
        
        try:
            with open("reg_users.json","r") as f:
                data=json.load(f)
        except json.decoder.JSONDecodeError as e:
            print("File Is Empty First Register Yourself!!!!")
            main_screen()
        except FileNotFoundError:
            print("File not Found First Register Yourself!!!")
            main_screen()
        e_m=input("Enter Your Registered Email")
        if e_m in list(data.keys()):
            for k,v in data[e_m].items():
                print(k,":",v)
        val1=input("Enter ther field to update : ").capitalize()
        
        if val1=="Email":
            print("For  Security reason you cannot edit this value")
        else:
            data[e_m][val1]=input("Enter the new Value: ")
        
        with open("reg_users.json","w") as f:
            json.dump(data,f)
        print("Successfully Update!!!!")
        x=input("********Press Enter To Go Back To Main Menu********")
        if x=="":
            self.user_menu()
    def order_history(self):
        file_name=self.email_f + ".json"
        try:
            with open(file_name,'r') as f:
                data=json.load(f)
        except:
            print("You Havent ordered yet")
        print(data)
        for k,v in data.items():
            print("Date of Order :",k)
            for item in v:
                
                print(f"{item[0].capitalize()} ({item[1]}) [INR{item[2]}]", end=' ')
                print()
        
        x=input("********Press Enter To Go Back To Main Menu********")
        if x=="":
            self.user_menu()
        

            
c=Admin()
d=User()
def main_screen():
    print("\n")
    print('---------------------------')
    print('|     (A)ADMIN LOGIN       |')
    print('---------------------------')
    print('---------------------------')
    print('|     (U)USER LOGIN       |')
    print('---------------------------')
    print('---------------------------')
    print('|     (R) REGISTER        |')
    print('---------------------------')
    while True:
        print("NOTE:  For Admin enter A ,for Registration Enter R ,for user Enter  U and to Exit Just Enter E")
        ch=input("Please Input Your Choice").upper()
        if (len(ch)==1):
            if ch=='A':
                c.admin_login()
                break
            elif ch=="R":
                d.reg_form()
                break
            elif ch=="U":
                d.user_login() 
                break
            elif ch=="E":
                print("*" * 32 + "THANK YOU" + "*" * 31 + "\n") 
                break
            else:                                                                                 
                print("\n" * 10 + "ERROR: Invalid Input . Try again!")   
        else:                                                                                     
            print("\n" * 10 + "ERROR: Invalid Input . Try again!")
main_screen()
```
